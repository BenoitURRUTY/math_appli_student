### 3. *Exercice 3 : description et modélisation de données*

Dans la première séance, on a utilisé l’approche ‘gaussienne’ pour décrire le jeu de données (de température), avec les métriques standards (moyenne, écart type) qui s’interprètent en terme de probabilité de distribution. Si ces données suivent bien la loi de distribution (normale) déduite de l’échantillon, alors, par exemple l’intervalle \[m‑σ,m+σ\] doit contenir 68% des mesures.

Une autre approche existe, beaucoup plus générale, pour décrire un jeu de données : les quantiles, qui sont les valeurs qui partitionnent le jeu en *n* groupes de même taille. Il s’agit de décrire la distribution empirique, sans chercher une loi particulière (approche dite “ non-paramétrique ”).

#### 3.1 *Robustesse de la médiane aux valeurs extrêmes*

Reprendre l’exercice sur les mesures de sulfate à Dumont d’Urville de la séance \#2 : comparer les moyennes avec et sans la valeur erronnée du 4/5/2006 (92000). Même chose avec la médiane.

<span style="color:blue">

Moyennes : 841 vs. 1218 (ng/ml), sans et avec la valeur erronée
Médianes : 589 vs. 590 , sans et avec la valeur erronée
Explications : la moyenne est très sensible aux valeurs extrêmes, car son calcul est basé sur les distances entre les valeurs. En effet, la moyenne m est égale à , de sorte que , m est le barycentre des valeurs.
Au contraire, la médiane est basée sur la répartition des valeurs. Enlever une valeur extrême fait ‘bouger’ la médiane d’une position, ce qui a peu d’impact sur la valeur de la médiane (en général). On dit que la médiane est ‘robuste’ par rapport aux valeurs extrêmes. Le schéma ci-dessous illustre cette propriété.
</span>

## Robustesse de la médiane

L'un des intérêts de la médiane est sa robustesse aux valeurs extrêmes, notamment lorsque le nombre de mesures est petit (en pratique, moins de 20). Le graphe ci-contre représente une série de mesures (à gauche), avec indiquées la moyenne (barre noire) et la médiane (barre grise). Si on supprime la valeur basse extrême (considérée comme ‘anormale’), à droite, la moyenne augmente sensiblement, mais la médiane ne change pas.
On voit donc que la médiane est d'un grand intérêt pour caractériser une série de valeurs; et aussi que la comparaison entre moyenne et médiane est intéressante.

```python
# Ici soit on charge le tableau de valeurs unicolonnes Argentiere_clip_vect
path_data_unicol = 'Argentiere_clip_vect.asc'
albedo = np.loadtxt(path_data, skiprows=6)

# Ou on enlève dans le script ces 0 et on transforme la matrice en vecteur 1D
albedo_unicol = albedo[np.where(albedo!=0)]

X_axis = np.linspace(1,1,len(albedo_unicol)) # On créé un vecteur uniquement composé de 1 et de la taille des valeurs d'albédos.
X_axis = np.ones(len(albedo_unicol)) # Variante pour créer ce vecteur ayant uniquement des valeurs = 1
fig = plt.figure(1)
plt.scatter(X_axis,albedo_unicol,marker='+')
plt.title('Dispersion des valeurs d\'albedo : Glacier d\'Argentiere le 23 Aout 2003') # \' permet d'écrire un apostrophe ' sans fermer la chaine de caractère).
plt.ylabel('Albedos')
#plt.savefig('Albedo_dispersion.pdf')    
```

Ici, le fait de tracer toutes les valeurs d’albedo sur une même abscisse les ordonne, et cela nous permet d'étudier la dispersion des valeurs. Comme montré en classe, on observe 3 groupes de valeurs. Un premier en dessous de ~0.35, un deuxième entre ~0.35 et ~0.5 et un dernier contenant des valeurs extrêmes au dessus de 0.5. Seul bémol dans cette représentation, impossible de savoir quelle classe contient le plus de points (c’est important car l'albédo est représentatif de l'état de surface dans notre cas : neige, glace, glace+débris?). Pour déterminer la ‘densité’ des valeurs, nous allons tracer les quartiles grâce au diagramme à Moustache. Pour rappel, on dénombre 3 quartiles dans un échantillon : 
- Q1, la valeur en dessous de laquelle se trouve 25 % des valeurs (en terme de nombre et non de valeur). Par exemple si nous avons 12 valeurs dans notre jeu de donnée, Q1 sera la 1/4 * 12 = 3ème valeur dans l'ordre croissant, peu importe sa valeur.
- Q2 , la valeur en dessous de laquelle se trouve 50 % des valeurs (aussi appelée Médiane!!!)
- Q3, la valeur en dessous de laquelle se trouve 75 % des valeurs

```python
fig = plt.figure(2)
plt.boxplot(albedo_unicol)
plt.ylabel('Albedos')
plt.title('Dispersion (Boxplot) des valeurs d\'albedo :\n Glacier d\'Argentiere le 23 Aout 2003') # J'ai rajouté ici \n. Cela signifie « NewLine » car sinon votre titre sera coupé lorsque vous enregistrerez le graphe. Cela permet de sauter de ligne.
#plt.savefig('Albedo_boxplot.pdf')
```

Ainsi, cette boite à moustache nous indique la valeur minimum (trait noir en bas), Q1 (début de la boite), Q2=Médiane (rouge), Q3 (fin boite) et le maximum (trait noir). Les croix noires sont les deux valeurs les plus élevées, considérées comme ‘anormales’ par rapport au reste des valeurs, et donc indiquées à part. Ceci nous permet d'avoir un bon aperçu de la distribution des valeurs. On calcule par habitude la moyenne d'un jeu de donnée sans trop savoir pourquoi (sensible aux valeurs extrêmes qui plus est...), ici l'approche des quantiles est aussi (voire plus?) importante. On peut dire avec ce diagramme qu'il y a un petit nombre de valeur avec de forts albedos (au dessus de Q3) alors que 50 % des valeurs de trouve en dessous de ~0.38. Comme vous pouvez le voir, Q2 n'est pas à égale distance de Q1 et de Q3 (alors qu'il y a le même effectif entre Q2 et Q3 et entre Q1 et Q2). Cela reflète une distribution non normale car la médiane (dans ce cas proche de la moyenne) n'est clairement pas centrée sur la distribution. 

```python
X_axis = np.linspace((1./111),1,num=111) # On crée l'axe normalisé. J'ai choisi de représenter cela entre 0 et 1 mais vous pouvez faire cela entre 0 et 100 %.
albedo_unicol_trie = np.sort(albedo_unicol) # Classe dans l'ordre croissant

fig = plt.figure(3)
plt.scatter(X_axis,albedo_unicol_trie,marker='_')
plt.ylabel('Albedos')
plt.xlabel('Index normalise')
plt.xlim([0,1])
plt.title('Dispersion des valeurs d\'albedo / index :\n Glacier d\'Argentiere le 23 Aout 2003')
#plt.savefig('Albedo_index.pdf')
```

On observe clairement une rupture de pente aux alentours de 0.32. Ainsi environ 45 % des pixels ont des valeurs inférieures à 0.32. Physiquement cette rupture représente la valeur d'albédo de transition entre la partie englacée et enneigée du glacier (ligne de neige dans le jargon glaciologique). En déduisant que 45 % du glacier était en glace, vous venez sans le savoir de calculer l'AAR (ratio entre aire d'accumulation et d'ablationd'un glacier) paramètre clé pour calculer la perte de masse d'un glacier ou pour mesurer l'impact de fonte d'une année sur l'autre … Vous pouvez même regarder sur la carte la zone où se trouve la limite glace/neige en traçant (avec les mains) la ligne d'iso-albédo = 0.32.

### Rappels de cours : métriques de corrélation par méthode quadratique (‘moindres carrés’)

Pour deux séries x et y de *n* éléments, de moyennes $\bar{x}$ et $\bar{y}$, on peut calculer les coefficients suivants :

les variances et $$Var_x=\frac{1}{n} \cdot \sum_1^n(x-\bar{x})^2 $$  $$Var_y=\frac{1}{n} \cdot \sum_1^n(y-\bar{y})^2$$,

les écarts types $ \sigma_x=\sqrt{Var_x}$ et $\sigma_y=\sqrt{Var_y} $,

la covariance $$Cov_{xy}=\frac{1}{n}\sum_1^n(x-\bar{x})\cdot(y-\bar{y})$$,

le coefficient de corrélation est alors $r=\frac{Cov_{xy}}{\sigma_x\sigma_y}$.

La régression linéaire est obtenue en minimisant la somme des distances quadratiques (SSE = *Sum of squared errors*) entre chaque point et la droite de régression d’équation y = ax + b : $$SSE=\sum_1^n(y-(ax+b))^2 $$. On peut alors déterminer analytiquement a et b comme : a = *r*. σ<sub>y</sub>/σ<sub>x</sub> et $b = \bar{y} - a.\bar{x}$.

On suppose que les séries x et y ont des moyennes qui vérifient effectivement $\bar{y}(x) = ax+b$, et que chaque valeur y est distribuée normalement autour de $\bar{y}(x)$, avec un écart type σ<sub>e</sub>. Un bon estimateur de σ<sub>e</sub> est alors : $$\sigma_e=\sqrt{\frac{1}{n-2}\cdot\sum_1^n(y_(ax+b))^2} $$.

Les incertitudes sur *a* et *b* peuvent être estimées par leur écart type : $$\sigma_a=\frac{\sigma_e}{\sqrt{n}\cdot\sigma_e}$$ et $$\sigma_b=\frac{\sigma_e}{\sqrt{n}\cdot\sigma_e}\cdot\sqrt{\frac{1}{n}\cdot\sum_1^nx^2}$$

 Intervalles de confiance :

De façon standard on définit les intervalles de confiance au niveau *p* (par exemple *p*=95%) avec la distribution de Student et le coefficient t<sub>n-2,(1-p)/2</sub> (n-2 degrés de liberté) c’est à dire la valeur de t qui laisse α = (1‑p)/2 probabilité résiduelle.

L’intervalle de confiance sur a est ±σ<sub>a</sub>.t<sub>n-2,α</sub> et sur b ±σ<sub>b</sub>.t<sub>n-2,α</sub> .

L’intervalle de confiance sur la droite de régression (c’est à dire sur la valeur moyenne y(x) ) est : 

$$y(x)\pm \sigma_e\cdot t_{n-2,\alpha}\sqrt{\frac{1}{n}+\frac{(x-\bar{x})^2}{\sum_1^n(x-\bar{x})^2}}=y(x)\pm \sigma_e\cdot t_{n-2,\alpha}\sqrt{\frac{1}{n}+\frac{(x-\bar{x})^2}{n\cdot\sigma_x^2}}$$

L’intervalle de confiance sur la distribution des valeurs y(x) autour de leur moyenne y(x) est : $$y(x)\pm \sigma_e\cdot t_{n-2,\alpha}\sqrt{1+\frac{1}{n}+\frac{(x-\bar{x})^2}{n\cdot\sigma_x^2}}$$
